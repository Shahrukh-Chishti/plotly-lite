```python
from plotly_lite import utils
from numpy import array,arange,random,sin,exp,cos
import plotly.offline as py
import plotly.graph_objs as go
```

```python
nX,nY = 100,150
X = arange(nX)
Y = arange(nY)
#Z = exp(sin(X/100))*sin(Y/150)**2
Z = utils.empty((nX,nY))
```

```python
for idX,x in enumerate(X):
    row = exp(sin(x/100))*sin(Y/150)**2*cos((-1**x))
    Z[idX] = row
```

```python
Z.shape
```

```python
utils.plotHeatmap(Z,X,Y)
```

```python
fig = go.Figure(data =
    go.Contour(
        z=Z,x=X,y=Y,contours_coloring='heatmap'
    ))
fig.show()
```
